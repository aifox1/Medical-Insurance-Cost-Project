# U.S. Medical Insurance Costs

```python
import csv
import matplotlib.pyplot as plt


age_list = []
sex_list = []
bmi_list = []
children_list = []
smoker_list = []
region_list = []
charges_list = []

with open('insurance.csv') as insurance_obj:
    insurance_dataset = csv.DictReader(insurance_obj)
    for row in insurance_dataset:
        age_list.append(row["age"])
        sex_list.append(row["sex"])
        bmi_list.append(row["bmi"])
        children_list.append(row["children"])
        smoker_list.append(row["smoker"])
        region_list.append(row["region"])
        charges_list.append(row["charges"])    

class PatientsDS:
    def __init__(self, age_list,sex_list,bmi_list,children_list,smoker_list,region_list,charges_list):
        self.ages = age_list
        self.sexes = sex_list
        self.bmi = bmi_list
        self.children = children_list
        self.smoker = smoker_list
        self.region = region_list
        self.charges = charges_list
        
    # 1. Calculate average age of the patients. 
    def average_age(self):
        total_age = 0
        for age in self.ages:
            total_age += int(age)
        return print("1. Average age of the patients is " + str(round(total_age/len(self.ages),2)) + " years.\n") 
 

    # 2. Analyze majority of the Patients are from.
    def region_majority(self):
        regions_dict = {}
        for region in self.region:
            if region in regions_dict:
                regions_dict[region] += 1
            else:
                regions_dict[region] = 1
        
        region_maj = max(regions_dict,key=regions_dict.get)
        print("2. Majority of the patients are from " + region_maj + " region.")
        
        region = regions_dict.keys()
        count = regions_dict.values()
        plt.xlabel('Region')
        plt.ylabel('Number of Patients')
        plt.bar(region,count)
        plt.show()

    
    # 3. Average Medical Cost between Male and Female.
    def gender_avg_cost(self):
        total_cost_male = 0
        total_cost_female = 0
        count_male = 0
        count_female = 0
        for index in range(len(self.sexes)):
            if self.sexes[index] == 'male':
               total_cost_male += float(self.charges[index])
               count_male += 1
            else:
                total_cost_female += float(self.charges[index])
                count_female += 1
                         
        avg_cost_male = total_cost_male/count_male 
        avg_cost_female = total_cost_female/count_female
        print("3. Average medical insurance cost for male is $" + str(round(avg_cost_male,2)) +" and female is $"+ str(round(avg_cost_female,2)))
        
        plt.xlabel('Gender')
        plt.ylabel('Medical Charges')
        plt.bar(['male','female'],[avg_cost_male,avg_cost_female])
        plt.show()

        
    #4. Analyze average age is for someone who has at least one child.
    def avg_age_has_children(self):
        total_age = 0
        count = 0
        for index in range(len(self.children)):
            if int(self.children[index]) > 0:
                total_age += int(self.ages[index])
                count += 1
        avg_age = total_age/count
        return print("4. Average age for someone who has at least one child is " + str(round(avg_age,2)) + " years.")
            
patients = PatientsDS(age_list,sex_list,bmi_list,children_list,smoker_list,region_list,charges_list)
patients.average_age()
patients.region_majority()
patients.gender_avg_cost()
patients.avg_age_has_children()
```
